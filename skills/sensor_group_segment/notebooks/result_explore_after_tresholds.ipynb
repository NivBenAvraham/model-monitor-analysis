```python
df = pd.read_parquet(r'/Users/nivbenavraham/dev/model-monitor-analysis/data/results/2026-04-09 sensor grp segment/results.parquet')
df['date'] = pd.to_datetime(df['date'])
group_status = pd.read_csv(r'/Users/nivbenavraham/dev/model-monitor-analysis/ground_truth/ground_truth_statuess_ca_2026.csv').rename(columns={'status': 'group_status'})
group_status['date'] = pd.to_datetime(group_status['date'])
df = df.merge(group_status, on=['group_id', 'date'], how='left')

df.head()
```

```python
per_day = df.groupby(['group_id','date','group_status','hive_size_bucket'])['status'].value_counts().reset_index(name='count')
per_day['pct_per_size_per_day'] = per_day['count'] / per_day.groupby(['group_id','date','hive_size_bucket'])['count'].transform('sum')
per_day.groupby(['group_status','hive_size_bucket','status'])['pct_per_size_per_day'].median().reset_index().round(3)
```
